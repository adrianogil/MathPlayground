# Solving a Stardew Valley Routing Problem with Graph Theory & Python

based on https://towardsdatascience.com/solving-a-stardew-valley-routing-problem-with-graph-theory-python-fd4471077b3a

```python
#Import libraries
import pandas as pd
import numpy as np
import math
from itertools import product
import networkx as nx
from matplotlib import pyplot as plt
%matplotlib inline

#Initialise empty graph G
G = nx.Graph()

#Create 120 vertex labels
vertices = np.arange(0, 120, 1)

#Reshape to emulate 12x10 crop patch
layers = vertices.reshape((10,12))

#Add vertices and edges to G
for i in range(12):
    for j in range(10):
        v = layers[j][i]
        
        #For each vertex, store vertex label and layer/column position 0-11
        G.add_node(v, layer=i)
        
        #For each vertex, store edge to each adjacent vertex
        #Right
        if i == 11:
            pass
        else:
            G.add_edge(v, v+1)

        #Left
        if i == 0:
            pass
        else:
            G.add_edge(v, v-1)

        #Up
        if v < 12:
            pass
        else:
            G.add_edge(v, v-12)

        #Down
        if v > 107:
            pass
        else:
            G.add_edge(v, v+12)

#Set vertex positioning to layers of straight lines
pos = nx.multipartite_layout(G, subset_key='layer')
        
#Plot graph
plt.figure(figsize=(15, 10))
nx.draw(G, pos, node_size=1000, node_color='#43C3FF', with_labels=True, font_color='white');
```
